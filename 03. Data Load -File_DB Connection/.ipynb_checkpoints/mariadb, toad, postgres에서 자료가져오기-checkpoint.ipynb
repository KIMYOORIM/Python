### 01. my sql (heidi sql ) 에서 불러오기

```python
# PANDAS패키지 불러오기
import pandas as pd
import pymysql
from sqlalchemy import create_engine

#DB 커넥션 열기
engine = create_engine('mysql+pymysql://kopo:kopo@192.168.110.111:3306/kopo')

#DB 테이블을 읽어 Data Frame에 변수 저장하기
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine)
selloutData
selloutData2 = pd.read_sql_query('SELECT * FROM mysql_hs', engine)
selloutData2
#컬럼해더 재정의
selloutData.columns = ['종선','수찬','형섭','유림']

#데이터 저장
resultname='mysql_yoorim'
selloutData.to_sql (resultname, engine, if_exists='replace', index= False)
```

### 02. Oracle 에서 데이터 불러오기 (Toad)

```python
# 바뀌는거 거의 없음. 엔진만 바뀜.
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://KOPO2019:kopo2019@192.168.110.112:1521/xe') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine) 

# 컬럼해더 재정의
selloutData.columns = ['REGIONID','PRODUCTGROUP','YEARWEEK','VOLUME']

#데이터VIEW
print(selloutData.head())

#데이터 저장
resultname = 'oracleresult'
selloutData.to_sql(resultname, engine, if_exists='replace', index=False)

```

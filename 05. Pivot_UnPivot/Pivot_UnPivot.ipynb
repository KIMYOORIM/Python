###  Pivot ( 행을 열로)

```python
import pandas as pd
```

```python
selloutData = pd.read_csv("../Python_Fintech/Python_ST_EX/dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutData.head()
```

```python
pivotResult = selloutData.\
pivot_table(index = ['REGIONID','PRODUCT'], 
            columns='YEARWEEK', values='QTY', 
            aggfunc="mean")
#pivotResult2 = pivotResult.reset_index() //unpivot하기전 그룹 풀어주기 
```

```python
pivotResult.head()
```

```python
pivotResult.to_csv("./pivot2.csv")
```

```python
pivotResult = pivotResult.reset_index()
```

###  UnPivot (열을 행으로)

```python
#피벗했던애를 원상태로 복귀. 기존있는 애로 변경. 기준을 삼을 데이터 컬럼이름. (동일해야함?) 변수이름(넣을), value = 넣을값
```

```python
unpivotResult = pd.melt(
    pivotResult, id_vars=['REGIONID', 'PRODUCT'], 
    var_name='YEARWEEK', value_name='QTY')
```

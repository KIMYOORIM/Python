#### 2 .CountPlot : 특정컬럼 값의 개수 확인

```python
tips = sns.load_dataset('tips') #seaborn 에 있는 tips라는 데이터를 가져온다
tips.head()
```

```python
sns.countplot(x="day", data = tips) #x축에 'day'데이터를 넣어서 보여줘라 
```

```python
#day에는 목금토일이 있고 그 수는 이러하다. 
```

#### 3. Distplot : 특정 값의 분포도 확인

```python
#어디에 분포되어있는지 확인하고 싶은지.
```

```python
sns.distplot(tips['total_bill'])
```

#### Jointplot : 2차원 실수형 데이터 시각화

```python
#연관된 데이터를 보고싶을 때 
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

```python
#kind = 'kde' #10달러 정도 쓰면 2달러정도의 팁을 내는구나 알 수 있다. 
sns.jointplot(x= 'total_bill', y = 'tip', kind = 'kde',data = tips)
```

#### Pairplot : groupby개념

```python
#모든데이터에 대한 시각화
# 성별로 그룹한 데이터를 시각화 해서 보여준다. 
# 그룹별 특성을가지고 시각화해서 보여준다. 성별컬럼 그룹바이 성별이남성 이면 파랑 여성이면 주황색으로 표시
# 모든 차트를 그룹컨셉 적용시켜 시각화. 
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue = 'sex')
```

####  barplot : 데이터의 평균 및 실제범위를 시각화

```python
#시간별로 얼마정도 쓰는지  런치보단 디너에 평균적으로 더 많이 지불을 한다. 검은 선은 표준편차를 의미한다. 
#평균적인 데이터 량을 확인 할 수 있다 .
```

```python
sns.barplot(x='time', y='total_bill', data = tips)
```

#### lmplot  -> 기울기선를 나타내준다. 

```python
#성별을 가지고 상관관계도 알려준다. hue는 그룹바이다. 
sns.lmplot(x='total_bill', y = 'tip', data = tips )
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'] )
```

```python
xList=list(range(1,11,1)) #x는 1부터 11까지 1씩 점핑해서 만들어라. 
y=[] #빈리스트생성.
```

```python
#xList
```

#### legend 설정

```python
#y리스트 선언 후 안에 들어갈 값은 반복문시키고 싶다. 들어갈 값에 대해서 for i in x 만큼 i**2가 실행이 되는거다. 
#i**2가 계속해서 할당이 되는것이다. 단순하게 한줄에 끝내고 싶을때 사용. 포문만 역전을 시켜 돌릴 수있다. 
```

```python
x=list(range(1,11,1))
y = [i**2 for i in x] #y라는 리스트를 만들어 준다음에 그 안에 포문을 돌리면서 i**2를 만들어준 ylist 안에 넣어준다.
y2 = [i**2.2 for i in x] #i의2.2제곱 
```

```python
#포문디버깅, 포문을 한줄로 써서 리스트에다가 넣어주겠다. 
```

```python
# 차트 꾸미기, legend
plt.plot(x,y,"go--", label="first") #처음에 얘가 그려지고 
plt.plot(x,y2,"bo--", label="second") #얘가 두번째 그려진다. 그다음 라벨이  붙음
plt.legend(loc=0) #참조하는 포인트 찍어주면 붙게된다. loc => 붙을 위치가 바뀐다.  best -> 기본. 비어있는공간에 알아서 넣어줌.
```

```python
#크롤링, 데이터분석이랑은 분야가 다르다.
```

```python
# 차트 꾸미기, legend 확인 한번더. loc
plt.plot(x,y,"go--", label="first") #처음에 얘가 그려지고 
plt.plot(x,y2,"bo--", label="second") #얘가 두번째 그려진다. 그다음 라벨이  붙음
plt.legend(loc=2) #참조하는 포인트 찍어주면 붙게된다. loc => 붙을 위치가 바뀐다.  best -> 기본. 비어있는공간에 알아서 넣어줌.
```

```python
#그냥 참고만. 화살표 표시가 가능하다. 
plt.plot(x,y,"go--", label="first")
plt.plot(x,y2,"bo--", label="second")
plt.legend(loc=0)

plt.annotate('point',xy=(2,3), xytext=(4,4),arrowprops=dict(color="blue", width=2))
```

```python
# 가장 많이 쓰이는 차트 꾸미기 (x,y 일경우 따로 해주어야 한다. )
plt.plot(x,y, 'b', lw=1.5) #차트 선색 및 크기설정
plt.plot(x,y, 'bo', ) #차트 내 'o' 표기
plt.grid(True) #그리드 표기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
```

#### Axis (축)

```python
#figure  = 큰도화지 #주황색 큰틀 Axes 1,2라는 두개를 담을 거고 그안의 axis라는 축을 만들거다?
```

####  figure

```python
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]
```

```python
#캔버스를 그릴건데 2행 1열 -> 한페이지에 위아래로 두칸이 생기잖아?
#즉 일행 2열로 하면 좌우로 두개가 들어가겠지 
#212 2행 1열의 두번째, 211
```

```python
fig = plt.figure(figsize=(10,5)) #figsize (도화지 크기) 10,5는 상대적인 크기. 
ax1 = fig.add_subplot(211) #여기서 윗부분의 영역이 생기는 것이다.  2행1열의1번째. 
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot') #까지 윗 영역 설정
ax2 = plt.subplot(212) #여기서 아래의 영역이 생기는 것. 2행1열의2번째
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value') #여기부터 아래 역역 설정
#판매량정보와 계절성지수가 일치하는구나, 프로모션 비율, 가격 좌우로 배치해서 비교할 수 있다. 
```

#### 이중축 활용하기 subplot

```python
#1의범위는 는 0-100 인데 2의범위가 0-5 사이인경우 왼쪽에는 1기준, 오른쪽은 2기준으로 그린다. 
```

```python
fig = plt.figure(figsize=(10,5)) # 1. 도화지만든다
ax1 = fig.add_subplot(1,1,1) # 1행1열에 하나 즉 하나만 만든다.
plt.plot(x,y, 'b--',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = ax1.twinx() #ax1과 동일한 x축을 사용하면서 그리겠다. 
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

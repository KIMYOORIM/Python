## 3번문제

```python
## url 변수 담기 
resp = requests.get('https://www.timeanddate.com/holidays/south-korea/')

resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html,'html.parser') 

# html 부분구조 가져오기
columns=bs.select('thead > tr.head > th.sep')
columns

columnlist=[]
# for each 사용하여  columnlist 에 어펜드 후  데이터 프레임 으로 변환
for column in columns:
    columnlist.append(column.text)

contents=bs.select('tbody > tr')
dfcontent=[]
alldfcontents=[]
for content in contents:
    tds=content.findAll("td")
    for td in tds:
        if(td.text=="None"):
            continue;
        dfcontent.append(td.text)
    alldfcontents.append(dfcontent)
    dfcontent=[]
df=pd.DataFrame(columns=columnlist, data=alldfcontents)
```

```python
df
```

```python
resultname = 'KOPO_PY_성한빈_FINAL3'

df.to_sql(resultname,SaveEngine,if_exists='replace',index=False)
```
